**Iterating (for loop)** - On List, Tuple, Set, Dict

**Membership Test (in , not in)**- On List,Tuple,Set,Dict

```python
#List
a = ['apple','mango','cherry']

print('mango' in a)
print('apple' not in a)
```

```python
#Tuple
t = (23,54,56,'Ram')

print(34 in t)
print('Ram' not in t)
```

```python
#Set
s = {'apple','Kiran','Bhavesh'}

print('Bhavesh' in s)
print('apple' not in s)
```

```python
#Dict

d = {
    'name':'Apple',
    'color':'red',
     'prize':20
}

print('color' in d)
print('apple' not in d)
```

**Indexing & Negative Indexing & Slicing** - On List,Tuple,Set,Dict

```python
#List
list1 = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]

print(list1[:])
print(list1[2:6])
print(list1[-5:-2])
print(list1[:-2])
```

```python
#Tuple
tuple1 = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(tuple1[:])
print(tuple1[3:6])
print(tuple1[-4:-2])
print(tuple1[:-4])
```

**Some built-in functions for Python lists**

```python
#an empty list
empty_list=[]
 
#a list of strings
str_list=['this', 'is', 'a', 'list']
 
# add an element to the end of the list
str_list.append('sanjivani')
print(str_list)
 
#insert an item at the defined index
str_list.insert(3,'COE')
print(str_list)
 
#to get the index of the first matched item
print(str_list.index('a'))
 
#to count number of a specific element in a list
print(str_list.count('is'))
 
#to reverse the order of a list
str_list.reverse()
print(str_list)
 
#to sort the list in ascending order
str_list.sort()
print(str_list)
```

**Built-in functions for Tuples in Python**

```python
#a string tuple
tuple1=('this','coe','c','tuple')
 
#len(tuple) gives total length of a tuple
print(len(tuple1))
 
#max(tuple) gives maximum element of a tuple
print(max(tuple1))
 
#min(tuple) gives minimum element of a tuple
print(min(tuple1))
 
#count(x) gives number of occurances of x in the tuple
print(tuple1.count('coe'))
 
#index(x) gives index of first occurances of x in the tuple
print(tuple1.index('c'))
```
